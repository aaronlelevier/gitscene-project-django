```python
from collections import defaultdict

counts = defaultdict(lambda:0)
for commit in data:
    author = commit['commit']['author']
    d = dateutil.parser.parse(author['date'])
    key = d.strftime('%d-%b-%y')
    counts[key] += 1

with open('/Users/aaron/Documents/github/gitscene-project-d3js/line-chart-2/data-2.csv', 'w') as csvfile:
    fieldnames = ['date', 'close']
    writer = csv.DictWriter(csvfile, fieldnames=fieldnames)

    # header
    writer.writeheader()

    # rows
    for date, count in counts.items():
        print(date, count)
        writer.writerow({'date': date, 'close': count})
```
